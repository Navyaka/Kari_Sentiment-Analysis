```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time

# Set up the Chrome WebDriver using WebDriver Manager
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

try:
    # Open the Amazon reviews page
    driver.get("https://www.amazon.com/Kari-Amruta-Patil/product-reviews/8172237103/ref=cm_cr_getr_d_paging_btm_next_3?ie=UTF8&reviewerType=all_reviews&pageNumber=3")

    # Allow some time for the page to load
    time.sleep(30)

    # Find and print all review elements
    reviews = driver.find_elements(By.CSS_SELECTOR, "div.review")
    for review in reviews:
        # Print the review text
        review_text = review.find_element(By.CSS_SELECTOR, "span.review-text-content").text
        print(review_text)
        rev.append(review_text)
        print("\n" + "-"*80 + "\n")
        

    # Close the browser
    driver.quit()
except Exception as e:
    print(e)
    driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time

# Set up the Chrome WebDriver using WebDriver Manager
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

try:
    # Open the Amazon reviews page
    driver.get("https://www.amazon.com/Kari-Amruta-Patil/product-reviews/8172237103/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews")

    # Initialize an empty list to store reviews
    all_reviews = []

    while True:
        # Allow some time for the page to load
        time.sleep(3)

        # Find and print all review elements
        reviews = driver.find_elements(By.CSS_SELECTOR, "div.review")
        for review in reviews:
            try:
                # Extract the review text
                review_text = review.find_element(By.CSS_SELECTOR, "span.review-text-content").text
                # Add the review to the list if it's not already in the list
                if review_text not in all_reviews:
                    all_reviews.append(review_text)
                    print(review_text)
                    print("\n" + "-"*80 + "\n")
            except Exception as e:
                print(f"Could not extract review text: {e}")

        # Check for duplicates
        if len(set(all_reviews)) != len(all_reviews):
            print("Duplicate reviews found. Stopping the process.")
            break

        # Wait for 5 seconds before checking for new reviews
        time.sleep(10)


    # Close the browser
    driver.quit()

except Exception as e:
    print(f"Error: {e}")
    driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time

# Set up the Chrome WebDriver using WebDriver Manager
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

try:
    # Open the Goodreads reviews page for the book
    driver.get("https://www.goodreads.com/book/show/33285989/reviews?reviewFilters={%22workId%22:%22kca://work/amzn1.gr.work.v1.jqA0BF-v6Vh8FNZ4qH5Kow%22,%22after%22:%22MjEwLDE0OTQ0ODk5OTE5MzE%22}")

    # Allow some time for the page to load
    time.sleep(5)

    # List of XPaths for review sections
    xpaths = [
        '/html/body/div[1]/div[2]/main/div[1]/div[2]/div[5]/div[3]/div[1]/article/section/section[2]/section/div/div',
        '/html/body/div[1]/div[2]/main/div[1]/div[2]/div[5]/div[3]/div[3]/article/section/section[2]/section/div/div'
        # Add more XPaths as needed for additional reviews
    ]

    # Initialize an empty list to store all review texts
    all_reviews = []

    for xpath in xpaths:
        try:
            # Find all review elements using the XPath
            review_elements = driver.find_elements(By.XPATH, xpath)

            for review_element in review_elements:
                try:
                    # Extract review text
                    review_text = review_element.find_element(By.CSS_SELECTOR, "span").text.strip()

                    # Append review text to the list
                    all_reviews.append(review_text)

                except Exception as e:
                    print(f"Error extracting review text: {e}")

        except Exception as e:
            print(f"Error finding review elements with XPath {xpath}: {e}")

    # Print all extracted reviews
    for idx, review in enumerate(all_reviews, start=1):
        print(f"Review {idx}: {review}\n" + "-"*80 + "\n")

    # Print total number of reviews extracted
    print(f"Total Reviews Extracted: {len(all_reviews)}")

    # Close the browser
    driver.quit()

except Exception as e:
    print(f"Error: {e}")
    driver.quit()
```
